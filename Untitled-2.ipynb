```python
# ==============================================================================
# VQR ANALYSIS: FITTING POLYNOMIALS TO x AND x² FUNCTIONS
# ==============================================================================

import numpy as np           
import matplotlib.pyplot as plt  
import math                  
import random               
from qiskit import QuantumCircuit        
from qiskit.quantum_info import Statevector  
from scipy.optimize import minimize     

# Set seeds for reproducible results
np.random.seed(42)
random.seed(42)

print("VQR Analysis: Polynomial Fitting to Simple Functions")
print("Target functions: f(x) = x and f(x) = x²")
print("Fitting with: a₀ + a₁x + a₂x² + a₃x³ (4 parameters)")
print("=" * 60)

# ==============================================================================
# QUANTUM INNER PRODUCT (Same as before)
# ==============================================================================
def inner_prod(vec1, vec2):
    """
    Quantum inner product estimation using interference.
    Returns similarity between -1 (opposite) and 1 (identical).
    """
    if len(vec1) != len(vec2):
        raise ValueError('Vectors must be same length')
    
    N = len(vec1)
    nqubits = math.ceil(np.log2(N))
    
    # Combine vectors in quantum superposition
    vec = np.concatenate((vec1, vec2)) / np.sqrt(2)
    
    # Create quantum circuit
    circ = QuantumCircuit(nqubits + 1)
    circ.initialize(vec, range(nqubits + 1))
    circ.h(nqubits)  # Hadamard creates interference
    
    # Extract inner product from quantum state
    state = Statevector.from_instruction(circ)
    real_amps = np.real(state.data)
    m_sum = sum(real_amps[l]**2 for l in range(N))
    
    return 2 * m_sum - 1

# ==============================================================================
# VQR PARAMETERS
# ==============================================================================
order = 3  # Fit with cubic polynomial: a₀ + a₁x + a₂x² + a₃x³
N = 8      # Number of data points

# ==============================================================================
# VQR COST FUNCTION
# ==============================================================================
def calculate_cost_function(parameters):
    """
    VQR cost function using quantum inner product.
    
    For polynomial targets like x or x², this should theoretically
    find the exact coefficients (e.g., [0,1,0,0] for f(x)=x).
    """
    global x, y, ynorm
    
    # Build polynomial ansatz: a₀ + a₁x + a₂x² + a₃x³
    ansatz = parameters[0] * np.ones_like(x)  # Constant term
    for i in range(1, len(parameters)):
        ansatz += parameters[i] * x**i  # Add x^i terms
    
    # Normalize for quantum computation
    ansatz_norm = np.linalg.norm(ansatz)
    if ansatz_norm == 0:
        return 1.0  # Avoid division by zero
    ansatz_unit = ansatz / ansatz_norm
    
    # Quantum similarity measurement
    similarity = inner_prod(y, ansatz_unit)
    scaled_similarity = ansatz_norm / ynorm * similarity
    
    # Cost function: 0 = perfect fit, higher = worse fit
    return (1 - scaled_similarity) ** 2

# ==============================================================================
# SIMPLE POLYNOMIAL DATA GENERATION
# ==============================================================================
def generate_polynomial_data(func, func_name, noise=0.05):
    """
    Generate data for simple polynomial functions.
    
    func: Function to sample (like lambda x: x or lambda x: x**2)
    func_name: Name for printing
    noise: Amount of random noise to add
    """
    # Use domain [0.5, 3] to avoid issues with x=0 for x² and get reasonable range
    x = np.linspace(0.5, 3.0, N)
    y_clean = func(x)
    
    # Add proportional noise (since polynomial values can vary widely)
    noise_scale = np.maximum(np.abs(y_clean) * noise, 0.01)  # Minimum noise level
    y_noisy = y_clean + np.random.uniform(-1, 1, N) * noise_scale
    
    print(f"\n{func_name} data generated:")
    print(f"  Domain: x ∈ [{x.min():.1f}, {x.max():.1f}]")
    print(f"  Clean range: y ∈ [{y_clean.min():.2f}, {y_clean.max():.2f}]")
    print(f"  Noisy range: y ∈ [{y_noisy.min():.2f}, {y_noisy.max():.2f}]")
    print(f"  Noise level: {noise*100}%")
    
    return x, y_noisy

# ==============================================================================
# OPTIMIZER TESTING
# ==============================================================================
def run_optimizers(cost_fn, x0):
    """Test all optimizers on the VQR cost function."""
    methods = ["BFGS", "COBYLA", "Nelder-Mead", "CG", "trust-constr"]
    results = {}
    
    print("  Testing optimizers:")
    for method in methods:
        try:
            result = minimize(cost_fn, x0=x0, method=method,
                            options={'maxiter': 200}, tol=1e-6)
            results[method] = result
            status = "✓" if result.success else "⚠"
            print(f"    {method:<12}: {status} (cost: {result.fun:.6f}, iter: {getattr(result, 'nit', '?')})")
        except Exception as e:
            print(f"    {method:<12}: ✗ Failed - {str(e)[:30]}")
            results[method] = None
    
    return results

# ==============================================================================
# UTILITY FUNCTIONS
# ==============================================================================
def evaluate_polynomial(xfit, coeffs):
    """Evaluate polynomial with given coefficients."""
    yfit = np.zeros_like(xfit)
    for i, a in enumerate(coeffs):
        yfit += a * xfit**i
    return yfit

def calculate_metrics(y_true, y_pred):
    """Calculate MSE and R-squared."""
    mse = np.mean((y_true - y_pred)**2)
    ss_res = np.sum((y_true - y_pred)**2)
    ss_tot = np.sum((y_true - np.mean(y_true))**2)
    r_squared = 1 - (ss_res / ss_tot) if ss_tot != 0 else 0
    return mse, r_squared

# ==============================================================================
# MAIN VQR ANALYSIS FOR POLYNOMIAL FUNCTIONS
# ==============================================================================
def run_vqr_for_polynomial(target_func, func_name, expected_coeffs):
    """
    Run complete VQR analysis for polynomial target functions.
    
    target_func: Function to fit (e.g., lambda x: x)
    func_name: Name for display
    expected_coeffs: What we expect to find (e.g., [0,1,0,0] for f(x)=x)
    """
    global x, y, ynorm
    
    print(f"\n{'='*20} VQR Analysis: {func_name} {'='*20}")
    
    # Step 1: Generate polynomial data
    x, y = generate_polynomial_data(target_func, func_name)
    y_original = y.copy()
    
    # Step 2: Normalize for quantum processing
    ynorm = np.linalg.norm(y)
    y = y / ynorm
    print(f"  Data normalized (||y|| = {ynorm:.3f})")
    
    # Step 3: Random initialization
    x0 = [random.uniform(-0.5, 0.5) for _ in range(order + 1)]
    print(f"  Starting guess: {[f'{c:.3f}' for c in x0]}")
    print(f"  Expected result: {expected_coeffs} (theoretical)")
    
    # Step 4: Run optimizers
    results = run_optimizers(calculate_cost_function, x0)
    
    # Step 5: Analysis and visualization
    xfit = np.linspace(np.min(x), np.max(x), 100)
    metrics = {}
    
    plt.figure(figsize=(15, 10))
    
    # Main fit plot
    plt.subplot(2, 3, 1)
    plt.scatter(x, y_original, color='black', s=60, label='Noisy Data', zorder=5)
    
    # Plot true function
    y_true_fit = target_func(xfit)
    plt.plot(xfit, y_true_fit, 'g:', linewidth=3, alpha=0.8, label=f'True {func_name}')
    
    # Plot VQR results
    successful_methods = []
    for method, result in results.items():
        if result is not None and result.success:
            coeffs = result.x
            yfit = evaluate_polynomial(xfit, coeffs)
            y_pred = evaluate_polynomial(x, coeffs)
            
            mse, r2 = calculate_metrics(y_original, y_pred)
            metrics[method] = {
                'MSE': mse,
                'R²': r2,
                'Cost': result.fun,
                'Coefficients': coeffs,
                'Iterations': getattr(result, 'nit', 'N/A')
            }
            
            plt.plot(xfit, yfit, linewidth=2, label=f"{method} (R²={r2:.4f})")
            successful_methods.append(method)
    
    # Classical polynomial fit baseline
    classical_coeffs = np.polyfit(x, y_original, order)[::-1]  # Reverse to get [a₀,a₁,a₂,a₃]
    y_classical_fit = evaluate_polynomial(xfit, classical_coeffs)
    y_classical_pred = evaluate_polynomial(x, classical_coeffs)
    mse_classical, r2_classical = calculate_metrics(y_original, y_classical_pred)
    
    metrics['Classical'] = {
        'MSE': mse_classical,
        'R²': r2_classical,
        'Cost': 'N/A',
        'Coefficients': classical_coeffs,
        'Iterations': 'Direct'
    }
    
    plt.plot(xfit, y_classical_fit, 'r--', linewidth=2.5, 
             label=f'Classical (R²={r2_classical:.4f})')
    
    plt.title(f'VQR Polynomial Fitting: {func_name}')
    plt.xlabel('x')
    plt.ylabel('y')
    plt.legend()
    plt.grid(True, alpha=0.3)
    
    # Residuals
    plt.subplot(2, 3, 2)
    for method in successful_methods:
        coeffs = metrics[method]['Coefficients']
        y_pred = evaluate_polynomial(x, coeffs)
        residuals = y_original - y_pred
        plt.scatter(x, residuals, label=f'{method}', alpha=0.8)
    
    # Classical residuals
    classical_residuals = y_original - y_classical_pred
    plt.scatter(x, classical_residuals, label='Classical', marker='x', s=80, color='red')
    
    plt.title('Residuals')
    plt.xlabel('x')
    plt.ylabel('y_true - y_pred')
    plt.axhline(y=0, color='black', linestyle='-', alpha=0.3)
    plt.legend()
    plt.grid(True, alpha=0.3)
    
    # Coefficient comparison
    plt.subplot(2, 3, 3)
    coeff_names = ['a₀', 'a₁', 'a₂', 'a₃']
    x_pos = np.arange(len(coeff_names))
    
    # Plot expected coefficients
    plt.bar(x_pos - 0.3, expected_coeffs, 0.2, label='Expected', alpha=0.7, color='green')
    
    # Plot each method's coefficients
    bar_width = 0.6 / len(successful_methods)
    for i, method in enumerate(successful_methods):
        coeffs = metrics[method]['Coefficients']
        plt.bar(x_pos - 0.1 + i*bar_width, coeffs, bar_width, 
                label=method, alpha=0.7)
    
    # Classical coefficients
    plt.bar(x_pos + 0.1, classical_coeffs, 0.2, label='Classical', alpha=0.7, color='red')
    
    plt.xlabel('Coefficient')
    plt.ylabel('Value')
    plt.title('Learned Coefficients')
    plt.xticks(x_pos, coeff_names)
    plt.legend()
    plt.grid(True, alpha=0.3)
    
    # Performance metrics
    plt.subplot(2, 3, 4)
    all_methods = list(metrics.keys())
    r2_values = [metrics[m]['R²'] for m in all_methods]
    colors = ['red' if m == 'Classical' else 'blue' for m in all_methods]
    
    bars = plt.bar(all_methods, r2_values, color=colors, alpha=0.7)
    plt.xlabel('Method')
    plt.ylabel('R² Score')
    plt.title('Performance Comparison')
    plt.xticks(rotation=45)
    plt.ylim(0, 1.05)
    
    # Add value labels
    for bar, r2 in zip(bars, r2_values):
        plt.text(bar.get_x() + bar.get_width()/2, bar.get_height() + 0.01, 
                f'{r2:.4f}', ha='center', va='bottom', fontsize=9)
    plt.grid(True, alpha=0.3)
    
    # Cost function values
    plt.subplot(2, 3, 5)
    cost_methods = [m for m in successful_methods if 'Cost' in metrics[m]]
    cost_values = [metrics[m]['Cost'] for m in cost_methods]
    
    if cost_methods:
        plt.bar(cost_methods, cost_values, alpha=0.7, color='purple')
        plt.xlabel('VQR Method')
        plt.ylabel('Final Cost')
        plt.title('VQR Cost Function Values')
        plt.xticks(rotation=45)
        plt.yscale('log')  # Log scale since costs can be very small
        plt.grid(True, alpha=0.3)
    
    # Coefficient error from expected
    plt.subplot(2, 3, 6)
    if successful_methods:
        coeff_errors = []
        for method in successful_methods:
            coeffs = metrics[method]['Coefficients']
            error = np.linalg.norm(np.array(coeffs) - np.array(expected_coeffs))
            coeff_errors.append(error)
        
        # Add classical error
        classical_error = np.linalg.norm(np.array(classical_coeffs) - np.array(expected_coeffs))
        
        plt.bar(successful_methods + ['Classical'], 
                coeff_errors + [classical_error],
                color=['blue']*len(successful_methods) + ['red'], alpha=0.7)
        plt.xlabel('Method')
        plt.ylabel('||coeffs - expected||')
        plt.title('Coefficient Error')
        plt.xticks(rotation=45)
        plt.yscale('log')
        plt.grid(True, alpha=0.3)
    
    plt.tight_layout()
    plt.show()
    
    # Print detailed results
    print(f"\n  Results Summary:")
    print("  " + "-"*80)
    print(f"  {'Method':<12} {'R²':<8} {'MSE':<10} {'Cost':<12} {'Coeff Error':<12} {'Iter'}")
    print("  " + "-"*80)
    
    # Sort by R²
    sorted_metrics = sorted(metrics.items(), key=lambda x: x[1]['R²'], reverse=True)
    
    for method, metric in sorted_metrics:
        coeffs = metric['Coefficients']
        coeff_error = np.linalg.norm(np.array(coeffs) - np.array(expected_coeffs))
        
        cost_str = f"{metric['Cost']:.2e}" if metric['Cost'] != 'N/A' else 'N/A'
        
        print(f"  {method:<12} {metric['R²']:<8.4f} {metric['MSE']:<10.2e} "
              f"{cost_str:<12} {coeff_error:<12.2e} {metric['Iterations']}")
    
    # Best method analysis
    best_method = sorted_metrics[0][0]
    best_coeffs = sorted_metrics[0][1]['Coefficients']
    
    print(f"\n  🏆 Best Method: {best_method}")
    print(f"      R² = {sorted_metrics[0][1]['R²']:.6f}")
    print(f"      Learned: {[f'{c:.4f}' for c in best_coeffs]}")
    print(f"      Expected: {expected_coeffs}")
    
    return metrics, best_method

# ==============================================================================
# RUN ANALYSIS FOR x AND x² FUNCTIONS
# ==============================================================================

print("Testing VQR on polynomial functions where we know the exact answer!")
print("\nThis is perfect for testing because:")
print("- We know exactly what coefficients should be found")  
print("- We can see which optimizer gets closest to the true answer")
print("- Polynomials should be 'easy' for polynomial fitting")

# Test 1: Linear function f(x) = x
# Expected coefficients: [0, 1, 0, 0] (just the x¹ term)
linear_metrics, linear_best = run_vqr_for_polynomial(
    lambda x: x,           # Target function  
    "f(x) = x",           # Display name
    [0, 1, 0, 0]          # Expected coefficients [a₀, a₁, a₂, a₃]
)

# Test 2: Quadratic function f(x) = x²  
# Expected coefficients: [0, 0, 1, 0] (just the x² term)
quadratic_metrics, quadratic_best = run_vqr_for_polynomial(
    lambda x: x**2,       # Target function
    "f(x) = x²",         # Display name  
    [0, 0, 1, 0]         # Expected coefficients [a₀, a₁, a₂, a₃]
)

# ==============================================================================
# FINAL SUMMARY
# ==============================================================================
print("\n" + "="*60)
print("POLYNOMIAL VQR SUMMARY")
print("="*60)

print(f"Best optimizer for f(x) = x:  {linear_best}")
print(f"Best optimizer for f(x) = x²: {quadratic_best}")

if linear_best == quadratic_best:
    print(f"\n🎯 Consistent Winner: {linear_best}")
    print("   This optimizer consistently handles polynomial VQR problems best!")
else:
    print(f"\n📊 Different winners for different polynomial complexities:")
    print(f"   Linear (x):     {linear_best}")
    print(f"   Quadratic (x²): {quadratic_best}")

print(f"\n💡 Key Insights:")
print("✓ These are 'perfect' test cases - we know the exact answer")
print("✓ Shows which optimizer can recover known polynomial coefficients") 
print("✓ Reveals how well VQR handles different polynomial structures")
print("✓ Tests quantum cost function on simple, well-understood problems")

print(f"\n🔬 Why This Matters:")
print("- If VQR can't fit polynomials to polynomials well, it has fundamental issues")
print("- Best-performing optimizer here likely works well on harder problems")  
print("- Shows whether quantum approach adds value over classical methods")
print("- Validates the VQR implementation with known ground truth")
```
